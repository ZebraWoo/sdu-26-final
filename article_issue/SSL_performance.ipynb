```python
import matplotlib.pyplot as plt
import numpy as np

# ---- ROSE 数据 ----
iterations = np.array([31500, 45000, 58500, 72000])
rose_auc = np.array([0.943, 0.951, 0.956, 0.962])

# ---- PanDerm、SwaVDerm 数据 ----
panderm_auc = 0.952
swavderm_auc = 0.938

# 为三个模型设置低饱和度颜色
color_rose = "#CD6F91"     # 玫瑰灰
color_panderm = "#F8B653"  # 低饱和红
color_swav = "#5D89B0"     # 低饱和蓝

fig = plt.figure(figsize=(4.3, 3.2), dpi=200)
plt.rcParams["font.family"] = "Times New Roman"

# ---- ROSE 折线 + 五角星 ----
plt.plot(
    iterations,
    rose_auc,
    color=color_rose,
    marker="*",
    markersize=12,
    linewidth=2,
    label="ROSE",
)

# ---- PanDerm 散点：圆形 ----
plt.scatter(
    [iterations.max(), iterations.max()],
    [panderm_auc, panderm_auc],
    color=color_panderm,
    marker="o",
    s=90,
    label="PanDerm"
)

# ---- SwaVDerm 散点：三角形 ----
plt.scatter(
    [iterations.max(), iterations.max()],
    [swavderm_auc, swavderm_auc],
    color=color_swav,
    marker="^",
    s=90,
    label="SwaVDerm"
)

# 坐标轴设置
plt.xlabel("Pretraining iterations", fontsize=12)
plt.ylabel("Average AUC across 8 datasets", fontsize=12)
plt.xticks(iterations, [f"{int(i/1000)}k" for i in iterations])
plt.ylim(0.93, 0.97)
plt.yticks(np.arange(0.93, 0.971, 0.01))   # 每 0.01 一个刻度
# 网格与图例
# plt.grid(alpha=0.25)
# plt.legend(fontsize=12)

plt.tight_layout()
fig.savefig("article_figures/auc_8_datasets.pdf",format="pdf")
plt.show()
```
